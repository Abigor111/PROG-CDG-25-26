### Exercício 1

```python
class Produto:
    def __init__(self, preco_unitario, desconto, quantidade, aumento):
        self.preco_unitario = preco_unitario
        self.desconto = desconto
        self.quantidade = quantidade
        self.aumento = aumento
    def print_dados(self):
        return f"O preco unitario e {self.preco_unitario}, o desconto e {self.desconto}, a quantidade e {self.quantidade}, o aumento {self.aumento.perc_aumento}"
    def valor_pagar(self):
        valor_pagar = self.preco_unitario * self.quantidade
        return f"O valor a pagar pelo produto e de {valor_pagar}"
    def valor_desconto(self):
        desconto = self.desconto / 100
        valor_pagar = self.preco_unitario * self.quantidade
        return f"O valor a pagar pelo produto com desconto e de {(valor_pagar) - (valor_pagar * desconto)}"
```

```python
class Aumento_Previsto:
    def __init__(self, perc_aumento):
        self.perc_aumento = perc_aumento
    def print_aumento(self):
        return f"O aumento previsto e de{self.perc_aumento}"
```

```python
aumento = Aumento_Previsto(10)
produto = Produto(11, 3, 14, aumento)
produto.print_dados()
```

```python
produto.valor_desconto()
```

```python
produto.valor_pagar()
```

### Exercício 3

```python
def ParOuImpar(num):
    if num % 2 == 0:
        return 1
    else:
        return 0
```

```python
num = int(input("Digite um numero: "))
soma = 0
while num != 0:
    if ParOuImpar(num) == 1:
        soma += num
    num = int(input("Digite um numero: "))
with open("ex3.txt", "w") as f:
    f.write(f"A soma dos numeros pares e {soma}")
```
