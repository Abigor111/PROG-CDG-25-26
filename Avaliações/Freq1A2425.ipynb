````python
#Execicio 3

n=int(input("Insira um numero inteiro para a soma dos seus digitos: "))

while n !=0:
    soma=0
    while n <= 1000:
        print("Numero invalido insira um numero superior ou igual a 1000. ")
        n=int(input("Insira um numero inteiro para a soma dos seus digitos: "))
    while n > 0:
        digito =n % 10
        soma+=digito
        n = n // 10
    print(soma)
    n=int(input("Insira um numero inteiro para a soma dos seus digitos: "))
print("O programa vai terminar.")
````
