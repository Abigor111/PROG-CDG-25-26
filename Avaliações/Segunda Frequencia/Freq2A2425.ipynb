### Exercicio 1
Contar o número de vogais em uma palavra fornecida pelo utilizador.

````python
def contar_vogais(palavra):
    vogais = "aeiou"
    num_vogais = 0
    for x in palavra:
        if x.lower() in vogais:
            num_vogais += 1
    return num_vogais
resposta = input("Digite uma palavra: ")
num_vogais = contar_vogais(resposta)
print(f"O numero de vogais na palavra {resposta} e {num_vogais}")
````

### Exercicio 2

Considera a seguinte lista de números:
`lista_orig=[3, 4, 5, 6 ,7, 8, 3]`
 - a) Faça uma nova lista cujo os valores sejam o triplo da lista original.
 - b) Calcula o valor maximo, media e o mais frequente.
 - c) Adicione o valor da média ao final da lista.
 

````python
lista_orig=[3, 4, 5, 6 ,7, 8, 3]
# Achar o Triplo
lista_triplo = [x*3 for x in lista_orig]
# Achar o valor maximo
valor_max = max(lista_orig)
# Achar a media
media = sum(lista_orig) / len(lista_orig)
# Achar valor mais frequente
def mais_freq(lista):
    mais_freq = lista.count(lista[0])
    valor_mais_freq = lista[0]
    for x in lista:
        if lista.count(x) > mais_freq:
            mais_freq = lista.count(x)
            valor_mais_freq = x
    return valor_mais_freq
mais_frequente = mais_freq(lista_orig)
print(f"O triplo da lista {lista_orig} e {lista_triplo}")
print(f"O valor maximo da lista {lista_orig} e {valor_max}")
print(f"A media da lista {lista_orig} e {media}")
print(f"O valor mais frequente da lista {lista_orig} e {mais_frequente}")
# Adicionar valor da media
lista_orig.append(media)
````

### Exercicio 3
Com a seguinte lista de clientes:

```
lista_clientes=[
("Joao Pais", 4100, "Porto"),

("Ana Gomes", 1200, "Lisboa"),

("Teresa Cunha", 5430, "Valpacos")]
``` 

- Indica o número de clientes.
- Guarda apenas os nomes dos clientes numa nova lista.
Exibe os resultados no final.

````python
lista_clientes=[
("Joao Pais", 4100, "Porto"),
("Ana Gomes", 1200, "Lisboa"),
("Teresa Cunha", 5430, "Valpacos")]
num_clients = len(lista_clientes)
nomes_clientes = [x[0] for x in lista_clientes]
print(f"A lista tem {num_clients} clientes")
print(f"Os nomes dos clientes sao {nomes_clientes}")
````
