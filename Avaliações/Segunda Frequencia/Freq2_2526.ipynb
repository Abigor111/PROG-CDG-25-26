### Exercicio 1

```python
def funcao1(palavra):
    tamanho = len(palavra)
    minusculas = palavra.lower()
    return tamanho, minusculas
palavra = input("Digite uma palavra: ")
tamanho, minusculas = funcao1(palavra)
print(f"A seguinta palavra {palavra} tem {tamanho} caracteres.")
print(f"A seguinta palavra {palavra} fica desta maneira '{minusculas}' em minusculas.")
```

### Exercicio 2

```python
valores = [15, 17, 8, 0, 12, 17, 19, 20]
meio = len(valores) // 2
valores.insert(meio, 9)
um_terco = [x/3 for x in valores]
impares = [x for x in valores if x%2 != 0]
valores.sort()
valores.pop(0)
media = sum(valores) / len(valores)
minimo = min(valores)
mais_freq = valores.count(valores[0])
valor_mais_freq = valores[0]
for x in valores:
    if valores.count(x) > mais_freq:
        mais_freq = valores.count(x)
        valor_mais_freq = x
print(f"A lista em que os valores sejam 1/3 da original fica da seguinte maneira: {um_terco}")
print(f"A lista so com os numeros impares fica da seguinta maneira: {impares}")
print(f"A media da lista {valores} e {media}")
print(f"O valor minimo da lista {valores} e {minimo}")
print(f"O valor mais frequente da lista {valores} e o {valor_mais_freq} que aparece {mais_freq} vezes na lista.")
```
