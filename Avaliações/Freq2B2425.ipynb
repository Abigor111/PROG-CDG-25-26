### Exercicio 2

```python
lista = [3, 4, 5, 6, 7, 8]
soma = sum(lista)
tamanho = len(lista)
numeros_frequentes = [lista[0]]
mais_freq = lista.count(lista[0])
for x in range(1, len(lista)):
    freq = lista.count(lista[x])
    if freq > mais_freq:
        mais_freq = freq
        numeros_frequentes = [lista[x]]
    elif freq == mais_freq:
        numeros_frequentes.append(lista[x])
superior_4 = [x for x in lista if x>4]
print(f"A soma dos valores da lista: {lista} e igual a {soma}")
print(f"O tamanho da lista {lista} e {tamanho}")
print(f"Os valores mais frequentes da {lista} sao {numeros_frequentes}")
print(f"Na lista {lista}, os valores superiores a 4 sao {superior_4}")
print(f"Se os valores da lista {lista} forem ordenadas, ela fica da seguinta maneira: ")
lista.sort()
print(lista)
```

### Exercicio 3

```python
print("Bem vindo ao restaurante X")
menus = {1: ("Café e bolo de arroz", 1.3), 
 2: ("Dois cafés e meia torrada", 2.2), 
 3: ("Meia de leite e tosta com manteiga", 3),
 4: ("Galão com tosta mista", 3.5)
                 }
for x, y in menus.items():
    print(f"{x}º Menu: ")
    print(f"{y[0]}")
    print(f"Preco: {y[1]}€")
print(f"Vamos iniciar o seu pedido...")
conta = 0
pedido = int(input("O que vai desejar pedir? Por favor indique apenas o numero do menu: "))
itens_pedidos = {}
while pedido < 1 or pedido > 4:
    pedido = int(input("Esse pedido nao existe, temos apenas 4 menus. Por favor escolha um deles: "))
while pedido != 0:
    while pedido < 1 or pedido > 4:
        pedido = int(input("Esse pedido nao existe, temos apenas 4 menus. Por favor escolha um deles: "))
    preco = menus[pedido][1]
    quantidade = int(input("Que quantidade vai desejar? "))
    while quantidade < 1:
        quantidade = int(input("Por favor, introduza uma quantidade valida: "))
    conta += preco * quantidade
    itens_pedidos[menus[pedido][0]] = quantidade
    pedido = int(input("Deseja pedir mais alguma coisa? Se sim por favor indique o numero do pedido. Caso contrario introduza 0: "))
print("O pedido foi o seguinte: ")
for x, y in itens_pedidos.items():
    print(f"{y} {x}")
print(f"A conta final ficou no valor de {conta}")
```
