### Exercicio 2

````python
# Objetivo do codigo: Criar uma calculadora
num1 = int(input("Digite um numero: ")) # Definição da variavel num1
num2 = int(input("Digite outro numero: ")) # Definição da variavel num2
operation = int(input("Selecione uma operacao (1 - Adicao, 2- Subtracao, 3-Multiplicao, 4- Divisao) ")) # Variavel que define se a operação vai ser adicao, substracao, multiplicao ou divisao
if operation == 1: # Como dito no enunciado se o numero inserido for 1, vai ser uma adicao
    print(num1+num2) # Da print à adicao dos numeros
elif operation == 2: # Como dito no enunciado se o numero inserido for 2, vai ser uma subtraca
    print(num1-num2) # Da print à subtracao dos numeros
elif operation == 3: # Como dito no enunciado se o numero inserido for 1, vai ser uma multiplicacao
    print(num1*num2) # Da print à multiplicacao dos numeros
elif operation == 4: # Como dito no enunciado se o numero inserido for 1, vai ser uma divsao
    print(num1/num2) # Da print à divsao dos numeros
````

### Exercicio 3

````python
# Objetivo do codigo achar os numeros primos entre 1 e N (numero escolhido pelo utilizador)
N = int(input("Digite um numero inteiro")) # Variavel escolhida pelo enunciado
# Loop entre 1 a valor da variavel anterior
for i in range(1, N+1): # Tem que ser n+1, para o N aparecer no loop
    divisores = 2 # Sabemos que todos os numeros tem divisores, exceto 1. Por isso já se assume que o numero vai ter 2 divisores
    for j in range(2, i): # Loop para fazer as divisoes entre 2 (não é por 1, porque sabemos que qualquer numero é divisivel por 1) e aqui nao é i+1 porque se nao o numero vai ser dividido por ele mesmo
        if i % j == 0: # O resto sendo igual a 0, significa que o numero é divisivel
            divisores +=1 # Adiciona 1 à variavel que indica o numero de divisores
    if divisores == 2 and i != 1: # Como o numero primo só tem 2 divisores e 1 não é um numero primo, o código abaixo só é executado se o numero for realmente primo
        print(f"{i} é primo!")
````
