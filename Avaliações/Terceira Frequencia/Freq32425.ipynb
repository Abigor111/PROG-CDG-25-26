### Exercicio 1

```python
class EstudanteISCAC:
    # Construtor da Classe
    def __init__(self, nome, apelido, tipo, curso, ano):
        self.nome = nome
        self.apelido = apelido
        self.tipo = tipo
        self.curso = curso
        self.ano = ano
    # Metodo para imprimir as informacoes do aluno
    def imprimir(self):
        print(f"Nome do aluno {self.nome} {self.apelido}")
        print(f"Tipo de Curso do aluno {self.tipo}")
        print(f"O aluno esta no curso {self.curso}")
        print(f"O aluno esta no ano {self.ano}")
class LCDG(EstudanteISCAC):
    #Construtor da Classe
    def __init__(self, nome, apelido, tipo, curso, ano, morada, localidade, ntelefone):
        super().__init__(nome, apelido, tipo, curso, ano)
        self.morada = morada
        self.localidade = localidade
        self.ntelefone = ntelefone
    def imprimir2(self):
        self.imprimir()
        print(f"O aluno vive em {self.morada}")
        print(f"O aluno reside na localidade {self.localidade}")
        print(f"O numero de telefone do aluno e o seguinte {self.ntelefone}")
aluno_igor = EstudanteISCAC("Igor", "Carvalheira", "Licenciatura", "CDG", 1)
aluno_igor.imprimir()
print("------Mais informacoes sobre o aluno Igor------")
aluno_igor2 = LCDG("Igor", "Carvalheira", "Licenciatura", "CDG", 1, "Rua das 3 Portas", "Kosovo", 915231041)
aluno_igor2.imprimir2()
```

### Exercicio 2

```python
def Valida(idade):
    try:
        idade = int(idade)
        if idade < 17 or idade > 77:
            idade = input("Introduza uma idade entre 17 a 77 anos: ")
            Valida(idade)
    except ValueError:
        print("A idade deve ser um numero inteiro")

idade = input("Introduz a tua idade: ")
Valida(idade)
with open("idade.txt", "w") as i:
    i.write(f"A idade do aluno e {idade}")
print("Codigo Concluido")
```
