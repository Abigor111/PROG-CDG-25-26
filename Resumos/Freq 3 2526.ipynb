## Ficheiros

### Modos de Ficheiros

|Modo | Letra| Descrição|
|---------|---------|---------|
|Read|"r"|Abre o ficheiro para leitura|
|Write|"w"|Abre o ficheiro para escrita (um novo ficheiro é criado ou é subscrito o existente)|
|Append|"a"|Abre o ficheiro para adicionar conteúdos no final do mesmo, sem subscreve-lo|
|Read and Write |"r+"|Abre o ficheiro para leitura e escrita|
|Write and Read|"w+"|Abre o ficheiro para escrita e leitura (um novo ficheiro é criado ou é subscrito o existente)|
|Append and Read|"a+"|Abre o ficheiro para leitura e adicionar conteúdos no final do mesmo, sem subscreve-lo|



### Métodos para Configuração de Ficheiros

| Método      | Descrição                                                                         |
| ----------- | --------------------------------------------------------------------------------- |
| `read()`  | Lê todo o conteúdo do ficheiro                                         |
| `readline()`   | Lê uma linha de cada vez do fichero                                              |
| `readlines()`    | Converte todas as linhas numa lista                                                        |
| `write()`   | Escreve uma string para o ficheiro                          |
| `writelines()`  | Escreve uma lista de string para o ficheiro |
| `close()`   | Fecha o Ficheiro                  |

## Módulos Standard

### Math

| Método      | Descrição                                                                         |
| ----------- | --------------------------------------------------------------------------------- |
| `sqrt()`  | Retorna a raiz quadrada do número                                     |
| `pow()`  | Retorna o valor de uma potência                                         |
| `sin()`  | Retorna o seno de um ângulo em radianos                                        |
| `cos()`   | Retorna o cosseno de um ângulo em radianos                                               |
| `add()`    | Soma dois números                                                      |
| `subtract()`   | Subtrai dois números                         |
| `pi()`  | Retorna o valor de pi |
| `fsum()`   | Retorna a soma dos produtos                  |

### Random

| Método      | Descrição                                                                         |
| ----------- | --------------------------------------------------------------------------------- |
| `random()`  | Cria um conjunto de valores aleatórios entre 0 e 1                                        |
| `randint()`   | Devolve um inteiro entre a e b(parametros da função)                                              |
| `choice()`    | Escolhe um item de aleatório de uma lista/tupla/string                                                        |
| `shuffle()`   | Embaralha os elementos de uma sequência, alterando a original                       |

### Os

| Método      | Descrição                                                                         |
| ----------- | --------------------------------------------------------------------------------- |
| `getcwd()`  | Devolve a diretoria atual                                       |
| `listdir()`   | Lista o conteúdo da diretoria especificada                                              |
| `mkdir()`    | Cria uma diretoria com o nome especificado                                                       |
| `rename()`   | Renomeia o nome da diretoria                        |
| `rmdir()`  | Remove a diretoria especificada |

## Erros

| Erro      | Explicação                                                                         |
| ----------- | --------------------------------------------------------------------------------- |
| <p style="color:red;">SyntaxError</p>  | Ocorre quando o Python encontra código que não está conforme com as regras da linguagem                                   |
| <p style="color:red">TypeError</p>   | Gerado quando uma função ou operação é aplicado a um tipo de valor inapropiado                                             |
| <p style="color:red">ValueError</p>    | Ocorre quando uma função recebe um argumento do tipo indicado mas com um valor inapropiado, tal como passar à função um caracter quando era esperado um inteiro                                                       |
| <p style="color:red">IndexError</p>   | Ocorre quando se tenta aceder a um indice que nao se encontra disponivel na lista ou tupla                       |
