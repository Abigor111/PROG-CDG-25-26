### Exercicio 1

```python
class FuncionarioEmpresa:
    # Construtor da Classe
    def __init__(self, id, nome, categoria, area, anos_servico):
        self.id = id
        self.nome = nome
        self.categoria = categoria
        self.area = area
        self.anos_servico = anos_servico

    def informacoes_funcionario(self):
        print(f"Numero do Funcionario: {self.id}")
        print(f"Nome do Funcionario: {self.nome}")
        print(f"Categoria do {self.nome}: {self.categoria}")
        print(f"Area do {self.nome}: {self.area}")
        print(f"Numero de anos de servico do {self.nome}: {self.nome}")

class FuncionarioDetalhado(FuncionarioEmpresa):
    # Construtor da Classe
    def __init__(self, id, nome, categoria, area, anos_servico, morada, cidade, contacto):
        super().__init__(id, nome, categoria, area, anos_servico)
        self.morada = morada
        self.cidade = cidade
        self.contacto = contacto

    def mostrar(self):
        self.informacoes_funcionario()
        print(f"Cidade do {self.nome}: {self.cidade}")
        print(f"Morada do {self.nome}: {self.morada}")
        print(f"Contacto do {self.nome}: {self.contacto}")
# Codigo Principal
jorgemelo = FuncionarioEmpresa(11, "Jorge Melo", "Junior", "Informatica", 2)
jorgemelo.informacoes_funcionario()

print("---Proximo Funcionario----")
ana_dias = FuncionarioDetalhado(3, "Ana Dias", "Senior", "Gestao", 14, "Rua da Ana Dias", "Anadia", 917230900)
ana_dias.mostrar()
```

### Exercicio 2

```python
def valida_nota(nota):
    try:
        nota = int(nota)
        if nota in range(0, 20):
            pass
        else:
            nota = input("Introduza uma nota entre 0 e 20")
            valida_nota(nota)
    except ValueError:
        print("A nota deve ser um numero inteiro")
# Codigo Principal
nota = input("Introduza uma nota entre 0 e 20")
valida_nota(nota)
with open("nota.txt", "w") as n:
    n.write(nota)
```
