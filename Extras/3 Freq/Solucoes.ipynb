### Exercicio 1

```python
class Funcionario:
    # Construtor da Classe (Alinea A e B)
    def __init__(self, id, nome, departamento, cargo, anos_experiencia):
        self.id = id
        self.nome = nome
        self.departamento = departamento
        self.cargo = cargo
        self.anos_experiencia = anos_experiencia
    # Alinea C
    def mostrar_info(self):
        print(f"Numero do Funcionario: {self.id}")
        print(f"Nome do Funcionario: {self.nome}")
        print(f"Departamento do {self.nome}: {self.departamento}")
        print(f"Cargo do {self.nome}: {self.cargo}")
        print(f"Numero de anos de servico do {self.nome}: {self.anos_experiencia}")
# Alinea D
class FuncionarioCompleto(Funcionario):
    # Construtor da Classe
    def __init__(self, id, nome, departamento, cargo, anos_experiencia, morada, cidade, contacto):
        super().__init__(id, nome, departamento, cargo, anos_experiencia)
        self.morada = morada
        self.cidade = cidade
        self.contacto = contacto

    def mostrar_info_completa(self):
        self.mostrar_info()
        print(f"Cidade do {self.nome}: {self.cidade}")
        print(f"Morada do {self.nome}: {self.morada}")
        print(f"Contacto do {self.nome}: {self.contacto}")

# Codigo Principal (Alinea E)
jorgemelo = Funcionario(11, "Jorge Melo", "Junior", "Informatica", 2)
jorgemelo.mostrar_info()

print("---Proximo Funcionario----")
ana_dias = FuncionarioCompleto(3, "Ana Dias", "Senior", "Gestao", 14, "Rua da Ana Dias", "Anadia", 917230900)
ana_dias.mostrar_info_completa()
```

### Exercicio 2

```python
def valida_nota(nota):
    # Alinea A
    try:
        nota = int(nota)
        # Alinea B
        if nota in range(0, 20):
            pass
        else:
            nota = input("Introduza uma nota entre 0 e 20")
            valida_nota(nota)
    except ValueError:
        print("A nota deve ser um numero inteiro")
# Codigo Principal (Alinea C)
nota = input("Introduza uma nota entre 0 e 20")
valida_nota(nota)
with open("txt/nota.txt", "w") as n:
    n.write(nota)
```

### Exercicio 3

```python
class Livro:
    # Construtor da Classe (alinea a)
    def __init__(self, titulo, autor, ano_publicacao, genero):
        self.titulo = titulo
        self.autor = autor
        self.ano_publicacao = ano_publicacao
        self.genero = genero
    # Alinea B
    def mostrar_info(self):
        print(f"Titulo do Livro: {self.titulo}")
        print(f"Autor: {self.autor}")
        print(f"Ano de Publicacao: {self.ano_publicacao}")
        print(f"Genero: {self.genero}")

# Alinea C
lista_livros = [Livro("No Longer Human", "Osamu Dazai", 1948, "Romance e Ficcao"), 
                Livro("Harry Poter e o Calice de Fogo", "JK Rowling", 2000, "Romance e Fantasia"),
                Livro("Omniscient Reader Viewpoint Volume 1", "Sing Shong", 2025, "Fantasia e Ficcao Apocaliptica")]
# Alinea D
def print_livros(lista):
    for livro in lista:
        print("---")
        livro.mostrar_info()
print_livros(lista_livros)
```

### Exercicio 4

```python
def valida_email(email):
    # Alinea A
    try:
        # Alinea B
        if not "@" in email:
            raise ValueError("O email nao possui @")
        
        utilizador, dominio = email.split("@") # Cria novas strings a partir do email
        utilizador = utilizador.strip(" ") # Remove espaços em branco no nome do utilizador
        
        if utilizador == "":
            raise ValueError("Nao foi fornecido nenhum utilizador")
        
        if not "." in dominio:
            raise ValueError("Dominio Invalido")
        return True
    
    except ValueError as erro:
        print(f"O seguinte erro foi encontrado  \n\n{erro}")

# Alinea C
email = input("Introduza um email: ")

if valida_email(email): # Apenas executa este codigo se a condiçao verdadeira for returnada da função
    with open("txt/emails.txt", "w") as e:
        e.write(email)

    print("Email escrito no ficheiro")
```

### Exercicio 5

```python
class Produto:
    # Alinea A
    def __init__(self, codigo, descricao, preco, stock):
        self.codigo = codigo
        self.descricao = descricao
        self.preco = preco
        self.stock = stock
    # Alinea B
    def __str__(self):
        return f"""
Codigo: {self.codigo}
Descricao: {self.descricao}
Preco: {self.preco}€
Quantidade em stock {self.stock}
    """
    # Alinea C
    def atualizar_stock(self, nova_quantidade):
        self.stock = self.stock + nova_quantidade

# Alinea D
guitarra = Produto(111, "Guitarra utilizada por David Gilmour", 251, 2)
guitarra.atualizar_stock(-1)
print(guitarra)

print("--- Novo Produto ---")

disco_vinil = Produto(3, "Album de 1970 dos Pink Floyd", 30, 8)
disco_vinil.atualizar_stock(3)
print(disco_vinil)
```

### Exercicio 6

```python
def valida_quantidade(quantidade):
    # Alinea A
    try:
        quantidade = int(quantidade)
        # Alinea B
        if quantidade <= 0:
            raise ValueError("Introduza um valor inteiro e positivo")
        return True
    except ValueError as erro:
        print(erro)

# Alinea C
quantidade = input("Introduza a quantide de artigos que vai comprar: ")
if valida_quantidade(quantidade):
    with open("quantidade.txt", "w") as q:
        q.write(quantidade)
    print("Quantidade registado no ficheiro")
```

### Exercicio 7

```python
# Alinea A
class Veiculo:
    def __init__(self, marca, modelo, ano):
        self.marca = marca
        self.modelo = modelo
        self.ano = ano
    def info(self):
        print(f"Marca: {self.marca}")
        print(f"Modelo: {self.modelo}")
        print(f"Ano: {self.ano}")
# Alinea B
class Carro(Veiculo):
    def __init__(self, marca, modelo, ano, portas):
        super().__init__(marca, modelo, ano)
        self.portas = portas
    def info_carro(self):
        self.info()
        print(f"Numero de Portas: {self.portas}")
        
class Mota(Veiculo):
    def __init__(self, marca, modelo, ano, cilindrada):
        super().__init__(marca, modelo, ano)
        self.cilindrada = cilindrada
    def info_mota(self):
        self.info()
        print(f"Cilindrada da Mota: {self.cilindrada}")

# Alinea D
autocarro = Veiculo("Honda", "BUS", 2011)
autocarro.info()

print("---")
toyota = Carro("Toyota", "CHR", 2021, 5)
toyota.info_carro()

print("---")
kawasaki = Mota("Kawasaki", "Z900", 2025, "muita")
kawasaki.info_mota()
```
