Exercicio 3.1.3

```python
def calculadora(num1, num2, operation):
    if operation == "+":
        def soma(num1, num2):
            return num1 + num2
        return soma(num1, num2)
    elif operation == "-":
        def sub(num1, num2):
            return num1 - num2
        return sub(num1, num2)
    elif operation == "*" or operation == "x":
        def mult(num1, num2):
            return num1 * num2
        return mult(num1, num2)
    elif operation == "/" or operation == ":":
        def div(num1, num2):
            return num1 / num2
        return div(num1, num2)
calculadora(1, 3, "+")
```

Exercicio 3.1.4

```python
def numero_primo(num):
    for i in range(2, num):
        if num % i == 0:
            break
    else:
        print(f"O número {num} é primo")
for i in range(10, 90):
    numero_primo(i)
```

Exercicio 3.1.5

```python
import random
def dado_100():
    um = 0
    dois = 0
    tres = 0
    quatro = 0
    cinco = 0
    seis = 0
    for i in range(1, 101):
        num = random.randint(1, 6)
        if num == 1:
            um += 1
        elif num == 2:
            dois += 1
        elif num == 3:
            tres += 1
        elif num == 4:
            quatro += 1
        elif num == 5:
            cinco += 1
        else:
            seis += 1
    print(f"O um foi lançado {um} vezes")
    print(f"O dois foi lançado {dois} vezes")
    print(f"O tres foi lançado {tres} vezes")
    print(f"O quatro foi lançado {quatro} vezes")
    print(f"O cinco foi lançado {cinco} vezes")
    print(f"O seis foi lançado {seis} vezes")
dado_100()
```

Exercicio 3.1.7

```python
def num_vogais(string):
    vogais = 0
    for i in range(0, len(string)):
        if string[i].lower() in ["a", "e", "i", "o", "u"]:
            vogais += 1
    print(vogais)
num_vogais("agua")
```

Exercicio 3.1.8

```python
def perfeito(num):
    soma_divisores = 0
    for j in range(1, num):
        if num % j == 0:
            soma_divisores += j
    if soma_divisores == num:
        return True
    else:
        return False
perfeito(28)
```

Exercicio 3.1.9

```python
def leibniz_pi(n):
    pi = 0
    for k in range(n):
        pi += (-1)**k / (2*k + 1)
    return 4 * pi
leibniz_pi(10)
```

Exercicio 3.1.12

```python
def mdc(num1, num2):
    for i in range(1, num1):
        if num1 % i == 0 and num2 % i == 0:
            maior_div_comum = i
    for i in range(1, num1):
        if num1 % i == 0 and num2 % i == 0:
            if i > maior_div_comum:
                maior_div_comum = i
    print(maior_div_comum)
mdc(30, 40)
```

Exercicio 3.1.15

```python
def fatorial_recurs(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * fatorial_recurs(n - 1)
```

Exercicio 3.1.16

```python
def fibonacci(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci(n-1) + fibonacci(n-2)
fibonacci(10)
```
