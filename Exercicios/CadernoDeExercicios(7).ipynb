### Exercicio 7.1

```python
class Carro:
    def __init__(self):
        self.marca = ""
        self.modelo = ""
        self.ano = ""
    def info(self):
        print("Marca", self.marca)
        print("Modelo", self.modelo)
        print("Ano", self.ano)
```

### Exercicio 7.2

```python
bmw = Carro()
bmw.marca = "BMW"
bmw.modelo = "Série 3"
bmw.ano = 2014
bmw.info()
audi = Carro()
audi.marca = "Audi"
audi.modelo = "BMW"
audi.ano = 2011
audi.info()
```

### Exercicio 7.3

```python
class CarrroDesportivo(Carro):
    def __init__(self):
        super().__init__()
        self.velocidade_maximo = ""
    def info(self):
        super().info()
        print(self.velocidade_maximo)
```

```python
audi = CarrroDesportivo()
audi.marca = "Audi"
audi.modelo = "BMW"
audi.ano = 2011
audi.velocidade_maximo = 111
audi.info()
```

### Exercicio 7.5

```python
class Motor:
    def __init__(self):
        self.cavalos = ""
        self.cilindradas = ""
    def ligar(self):
        print("O motor com a características")
        print("Cavalos", self.cavalos)
        print("Cilindrada", self.cilindradas)
```

```python
motor = Motor()
motor.cavalos = 111
motor.cilindradas = 111
motor.ligar()
```

### Exercicio 7.6

```python
class Carro:
    def __init__(self, marca, modelo, ano, motor):
        self.marca = ""
        self.modelo = ""
        self.ano = ""
        self.motor = motor
    def info(self):
        print("Marca", self.marca)
        print("Modelo", self.modelo)
        print("Ano", self.ano)
    def ligar(self):
        self.motor.ligar()
```

```python
motor = Motor()
motor.cavalos = 111
motor.cilindradas = 111
carro = Carro("audi", "a3", 2021, motor)
carro.ligar()
```
