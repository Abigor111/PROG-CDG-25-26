### Exercicios Extras com OOP

```python
class EstudanteCD:
    def __init__(self, nome, numero, teste1, teste2, teste3):
        self.nome = nome
        self.numero = numero
        self.teste1 = teste1
        self.teste2 = teste2
        self.teste3 = teste3
    def ClassFinal(self):
        if self.teste1 < 8 or self.teste2 < 8 or self.teste3 < 8:
            self.classificacao_final = 9
        else:
            testes = [self.teste1, self.teste2, self.teste3]
            media = sum(testes)/3
            self.classificacao_final = media

    def observacoes(self):
        if self.classificacao_final >= 10:
            print("Aprovado")
        elif self.classificacao_final == 9:
            print("Reprovado - Nao obteve os minimos")
        elif self.classificacao_final < 10:
            print("Reprovado")
    def info(self):
        print(f"O nome do aluno e {self.nome}")
        print(f"O numero do aluno e {self.numero}")
        print(f"A nota do primeiro teste do aluno e {self.teste1}")
        print(f"A nota do segundo teste do aluno {self.teste2}")
        print(f"A nota do terceiro teste do aluno {self.teste3}")
        print(f"A nota final do aluno {self.classificacao_final}")

igor = EstudanteCD("Igor", 111, 11, 11, 11)
igor.ClassFinal()
igor.info()
igor.observacoes()
```
