### **Divisão do Código Postal**
Desenvolva um programa que solicite ao utilizador um código postal no formato português, composto por quatro dígitos, seguidos de um hífen e três dígitos (ex.: 3111-103).

O programa deve:

- Ler o código postal introduzido pelo utilizador.

- Dividir o código postal em duas partes:

    + A primeira parte (antes do hífen), por exemplo 3111;

    + A segunda parte (após o hífen), por exemplo 103.

- Armazenar cada parte em variáveis distintas.

- Apresentar ao utilizador as duas variáveis separadamente.

```python
codigo_postal = input("Introduza o seu codigo postal (xxxx-yyy): ")
localidade, parte2 = codigo_postal.split("-")
print(f"O codigo referente a localidade do e {localidade}")
print(f"O codigo referente a outra parte {parte2}")
```
