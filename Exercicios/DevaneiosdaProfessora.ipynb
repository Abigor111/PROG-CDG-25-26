### **Divisão do Código Postal**
Desenvolva um programa que solicite ao utilizador um código postal no formato português, composto por quatro dígitos, seguidos de um hífen e três dígitos (ex.: 3111-103).

O programa deve:

- Ler o código postal introduzido pelo utilizador.

- Dividir o código postal em duas partes:

    + A primeira parte (antes do hífen), por exemplo 3111;

    + A segunda parte (após o hífen), por exemplo 103.

- Armazenar cada parte em variáveis distintas.

- Apresentar ao utilizador as duas variáveis separadamente.

````python
codigo_postal = input("Introduza o seu codigo postal (xxxx-yyy): ")
localidade, parte2 = codigo_postal.split("-")
print(f"O codigo referente a localidade do e {localidade}")
print(f"O codigo referente a outra parte {parte2}")
````

**Exercício: Cálculo de Idade e Verificação de Desconto com Data Específica**

Crie um programa em Python que solicita ao utilizador a sua data de nascimento no formato **dd/mm/aaaa** e verifica se o utilizador tem direito a desconto com base na sua idade, considerando que a data de corte para o desconto é **21 de novembro de 2025**. O programa deve:

1. Solicitar ao utilizador a sua **data de nascimento**.
2. Dividir a data em **dia**, **mês** e **ano**.
3. Calcular a **idade** do utilizador, considerando que o ano atual é **2025**.
4. Verificar se o utilizador tem **mais de 50 anos** e se ele já passou de **21 de novembro de 2025**. Caso contrário, ele terá direito ao desconto.
5. Exibir a mensagem:

   * "Tem direito a desconto" se o utilizador cumprir os critérios de idade e data.
   * "Não tem direito a desconto" caso contrário.

**Exemplo de entrada:**

```
Introduza a sua data de nascimento com o seguinte formato (dd/mm/aaaa): 15/05/1975
```

**Exemplo de saída:**

```
O dia é 15
O mês é 5
O ano é 1975
Tem direito a desconto
```

````python
data = input("Introduza a sua data de nascimento com o seguinte formato (dd/mm/aaaa): ")
dia, mes, ano = data.split("/")
dia = int(dia)
mes = int(mes)
ano = int(ano)
print(f"O dia e {dia}")
print(f"O mes e {mes}")
print(f"O ano e {ano}")
ano_atual = 2025
idade = ano_atual-ano
if idade > 50:
    if mes > 11:
        print("Nao tem direito a desconto")
    elif mes == 11:
        if dia > 21:
            print("Nao tem direito a desconto")
        else:
            print("Tem direito a desconto")
    else:  
        print("Tem direito a desconto")
else:
    print("Nao tem direito a desconto")
````

**Exercício: Verificação de Desconto com Base na Morada**

Crie um programa em Python que solicite ao utilizador que insira a sua morada no formato:

**Rua, Número, Andar, Localidade**

**Aviso:** Se o utilizador não viver num apartamento, deve introduzir **0** no campo de andar.

O programa deve:

1. Dividir a morada fornecida nos quatro componentes: **rua**, **número**, **andar** e **localidade**.
2. Mostrar cada uma das partes da morada.
3. Verificar se a **localidade** é **Coimbra**. Caso seja, o programa deve indicar que o utilizador tem direito a desconto. Caso contrário, deve indicar que o utilizador não tem direito a desconto.

**Exemplo de entrada:**

```
Rua das Flores, 456, 3, Coimbra
```

**Exemplo de saída:**

```
Rua: Rua das Flores
Número: 456
Andar: 3
Localidade: Coimbra
Tem direito a desconto!
```

````python
print("Siga o seguinte formato para introduzir a morada")
print("Rua, Numero, Andar, Local")
morada = input("Introduza a sua morada (Se nao viver num apartamento, meta 0 no andar): ")
rua, numero, andar, local = morada.split(",")
local = local.strip(" ")
print(f"Rua: {rua}")
print(f"Numero: {numero}")
print(f"Andar: {andar}")
print(f"Localidade: {local}")
if local == "Coimbra":
    print("Tem direito a desconto!")
else:
    print("Nao tem direito a desconto!")
````
