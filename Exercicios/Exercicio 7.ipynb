### Exercicio 7.1

```python
class Carro:
    def __init__(self, marca, modelo, ano, cor):
        self.marca = marca
        self.modelo = modelo
        self.ano = ano
        self.cor = cor
    def info(self):
        print("Marca", self.marca)
        print("Modelo", self.modelo)
        print("Ano", self.ano)
        print("Cor", self.cor)
```

### Exercicio 7.3

```python
class CarrroDesportivo(Carro):
    def __init__(self, marca, modelo, ano, cor, velocidade_maxima):
        self.marca = marca
        self.modelo = modelo
        self.ano = ano
        self.cor = cor
        self.velocidade_maxima = velocidade_maxima
    def info_desportivo(self):
        self.info()
        print(self.velocidade_maxima)
```

### Exercicio 7.5

```python
class Motor:
    def __init__(self, cavalos, cilindradas, ligar):
        self.cavalos = cavalos
        self.cilindradas = cilindradas
        self.ligar = ligar
    def ligar_motor(self):
        if self.ligar == 1:
            print("O carro ja esta ligado")
        else:
            self.ligar = 1
            print("O carro esta a ser ligado")
```

```python
motor = Motor(111, 111, 0)
motor.ligar_motor()
```

### Exercicio 7.6

```python
class Carro:
    def __init__(self, marca, modelo, ano, motor, cor):
        self.marca = marca
        self.modelo = modelo
        self.ano = ano
        self.motor = motor
        self.cor = cor
    def info(self):
        print("Marca", self.marca)
        print("Modelo", self.modelo)
        print("Ano", self.ano)
        print("Cor", self.cor)

    def ligar_carro(self):
        self.motor.ligar_motor()
```

```python
motor = Motor(111, 111, 0)
carro = Carro("audi", "a3", 2021, motor, "Red")
carro.ligar_carro()
```

```python
class Carro:
    def __init__(self, marca, modelo, ano, motor, cor):
        self.marca = marca
        self.modelo = modelo
        self.ano = ano
        self.motor = motor
        self.cor = cor
    def info(self):
        print("Marca", self.marca)
        print("Modelo", self.modelo)
        print("Ano", self.ano)
        print("Cor", self.cor)

    def ligar_carro(self):
        self.motor.ligar_motor()
class CarrroDesportivo(Carro):
    def __init__(self, marca, modelo, ano, cor, velocidade_maxima):
        self.marca = marca
        self.modelo = modelo
        self.ano = ano
        self.cor = cor
        self.velocidade_maxima = velocidade_maxima
    def info_desportivo(self):
        self.info()
        print(self.velocidade_maxima)
class Motor:
    def __init__(self, cavalos, cilindradas, ligar):
        self.cavalos = cavalos
        self.cilindradas = cilindradas
        self.ligar = ligar
    def ligar_motor(self):
        if self.ligar == 1:
            print("O carro ja esta ligado")
        else:
            self.ligar = 1
            print("O carro esta a ser ligado")
def exibir_informações(carro):
    carro.info()
def ligar_carro(carro):
    carro.ligar_carro()
motor = Motor(111, 111, 0)
carro = Carro("audi", "a3", 2021, motor, "Red")
carro.ligar_carro()
ligar_carro(carro)
exibir_informações(audi)
carro.info()
```
