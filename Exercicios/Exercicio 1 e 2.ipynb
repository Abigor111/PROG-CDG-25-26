Exercicico 2.2.28

```python
num_rom = input("Introduz um numero romano: (I, V, X, L, C, D, M)").upper()
num_arabe = 0
for i in range(0, len(num_rom)):
    if num_rom[i] == "I":
        num_arabe += 1
    elif num_rom[i] == "V":
        num_arabe += 5
        if num_rom[i-1] == "I":
            num_arabe -= 2
    elif num_rom[i] == "X":
        num_arabe += 10
        if num_rom[i-1] == "I":
            num_arabe -= 2
    elif num_rom[i] == "L":
        num_arabe += 50
        if num_rom[i-1] == "I":
            num_arabe -= 2
        elif num_rom[i-1] == "X":
            num_arabe -= 20
    elif num_rom[i] == "C":
        num_arabe += 100
        if num_rom[i-1] == "X":
            num_arabe -= 20
    elif num_rom[i] == "D":
        num_arabe += 500
        if num_rom[i-1] == "C":
            num_arabe -= 200
    elif num_rom[i] == "M":
        num_arabe += 1000
    else:
        print(f"{num_rom[i]} não pertence à numeração romana!")
print(f"{num_rom} em numeração arábe é igual a {num_arabe}!")
```

Exercicio 2.2.29

```python
num_arabe = int(input("Introduz um numero: "))
valores = [
    (1000, 'M'), (900, 'CM'), (500, 'D'), (400, 'CD'),
    (100, 'C'), (90, 'XC'), (50, 'L'), (40, 'XL'),
    (10, 'X'), (9, 'IX'), (5, 'V'), (4, 'IV'), (1, 'I')
]
resultado = ''
num = num_arabe
for valor, simbolo in valores:
    while num >= valor:
        resultado += simbolo
        num -= valor
print(f"{num_arabe} em numeração romana é igual a {resultado}!")
```

Exercicio 2.2.32

```python
nif = input("Introduza um numero nif válido: ")
soma_prod = 0
if len(nif) == 9:
    if int(nif[0]) in [1, 2, 3, 5, 6, 8 , 9]:
        for i in range(9, 1, -1):
            soma_prod += i * int(nif[-i])
        resto = soma_prod % 11
        if resto < 2:
            ultimo_dig = 0
        elif resto >= 2:
            ultimo_dig = 11- resto
        if ultimo_dig == int(nif[-1]):
            print(f"{nif} é válido!")
        else:
            print(f"{nif} é inválido!")
    else:
        print(f"{nif} é inválido!")
else:
    print(f"{nif} é inválido!")
```
