```python
# Exercicio 12
def numero_primo(lista):
    numeros_primos = []
    for num in range(0, len(lista)):
        for i in range(2, lista[num]):
            if lista[num] % i == 0:
                break
        else:
            numeros_primos.append(lista[num])
    return numeros_primos
lista = numero_primo([10, 11, 3, 2])
print(lista)
```

```python
# Exercicio 13
def max_ocurrence(lista):
    maior = lista.count(lista[0])
    idx_maior = 0
    for i in range(1, len(lista)):
        if maior < lista.count(lista[i]):
            maior = lista.count(lista[i])
            idx_maior = i
    return lista[idx_maior]
lista = max_ocurrence([10, 11, 3, 2, 2])
print(lista)
```

```python
# Exercicio 14
def list_max_ocurrence(lista):
    return max(lista), lista.count(max(lista))
lista = list_max_ocurrence([10, 11, 3, 2, 2])
lista
```
